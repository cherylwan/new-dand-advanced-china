# Cleaning 清洗

```python
# 合并3个表格
df_clean=pd.DataFrame()
df_clean=pd.merge(df,tweet_df,how='left',left_on='tweet_id',right_on='id')
```

```python
df_clean=pd.merge(df_clean, image_df, how='left',on='tweet_id')
```

```python
df_clean.info()
```

```python
# 删除没有图片的行
df_clean=df_clean.dropna(subset=['jpg_url'])
```

```python
# 删除属于转发的行

df_clean=df_clean.query('retweeted_status_id=="NaN"')
```

```python
df_clean.info()
```

```python
# 删除无用字段
df_clean.drop(['id','retweeted_status_id','retweeted_status_user_id','retweeted_status_timestamp'],axis=1,inplace=True)
```

```python
df_clean.head(5)
```

```python
breed=[]
conf=[]
pre_dog=[]


def prediction(dataframe):
    if dataframe['img_num']==1:
        breed.append(dataframe['p1'])
        conf.append(dataframe['p1_conf'])
        pre_dog.append(dataframe['p1_dog'])
    elif dataframe['img_num']==2:
        breed.append(dataframe['p2'])
        conf.append(dataframe['p2_conf'])
        pre_dog.append(dataframe['p2_dog'])
    elif dataframe['img_num']==3:
        breed.append(dataframe['p3'])
        conf.append(dataframe['p3_conf'])
        pre_dog.append(dataframe['p3_dog'])
    else:
        breed.append("NaN")
        conf.append("NaN")
        pre_dog.append("NaN")        

df_clean.apply(prediction, axis=1)

        
```

```python
#检查pre_dog项是否含有false，也就是最佳预测结果不属于狗狗图片
False in pre_dog
```

```python
df_clean.jpg_url[0]
```

```python
# 将breed...补充进dataframe
df_clean['breed']=breed
df_clean['conf']=conf
df_clean['pre_dog']=pre_dog
```

```python
# 删除img_num至p3_dog列
df_clean.drop(['img_num','p1','p1_conf','p1_dog','p2','p2_conf','p2_dog','p3','p3_conf','p3_dog'],axis=1,inplace=True)
```

```python
df_clean.info()
```

```python
# 狗狗stage列标明的数量
(df_clean.doggo!='None').sum()+(df_clean.floofer!='None').sum()+(df_clean.pupper!='None').sum()+(df_clean.puppo!='None').sum()
```

```python
#从text中提取狗狗所处的stage

df_clean['stage']=df_clean.text.str.extract('(puppo|pupper|floofer|doggo|floof)', expand=True)
df_clean['stage']=df_clean['stage'].replace('floof','floofer')
```

```python
df_clean.info()
```

```python
# 删除doggo等四列
df_clean.drop(['doggo','floofer','pupper','puppo'],axis=1,inplace=True)
```

```python
df_clean.info()
```

```python
df_clean=df_clean.dropna(subset=['in_reply_to_user_id'])
```

```python
#修改rating
df_clean[['numerator','denominator']]=df_clean.text.str.extract('(\d+\.?\d*)/(\d+\.?\d*)', expand=True)
```

```python
df_clean['numerator'].value_counts()
```

```python
#因为存在分子高于分母的情况，因此可以用分子除以分母，得到的商>=1均可认为是满分
df_clean['rating']=df_clean['numerator'].astype('float')/df_clean['denominator'].astype('float')
```
